```python
import pandas as pd

# The data is wrapped with this notebook.
# This is a smaller version of the 1990 census, available at UCI and other places.
# Taking a subset helps running time and to see the dendrogram.
census_df = pd.read_csv('USCensus1990-2000rows.data.txt') #read the data
census_df
```

```python
census_data = census_df.iloc[:, 1:] #do not include caseid in the data
```

```python
# While the implementation in sklearn is good,
# the scipy version can produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
%matplotlib inline

# z is the clustered data
Z = linkage(census_data, 'ward') # use Sum of squared distances between all observations

# plot the data
# This graph shows the hierarchical clustering of all the data points
# When the height is larger than 200, the data points can be cluster to two group
# when the height is larger than 100, all data points can be cluster to four group
# As the heigh is lower, more cluster will be formed.
plt.figure()
dendrogram(Z)
plt.show()
```

```python
NIPS_df = pd.read_csv('NIPS_1987-2015.csv') #read the data
NIPS_df
```

```python
NIPS_data = NIPS_df.iloc[:, 1:] #do not include word name in the data

Z = linkage(NIPS_data, 'ward') #z is the clustered data

# plot the data
# This graph shows the hierarchical clustering of all the data points
# When the height is larger than 300, the data points can be cluster to two group
# when the height is larger than 200, all data points can be cluster to four group
# As the heigh is lower, more cluster will be formed.
plt.figure(figsize=(10, 10))
dendrogram(Z)
plt.show()
```
